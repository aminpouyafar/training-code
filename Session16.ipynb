```python
import requests
```

```python
url = "https://www.iranjib.ir/"
response = requests.get(url) # یعنی خواندن و گرفتن دیتا
print(response.text) 
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(response.text, 'html.parser') #کل صفحه سایت
headline_bs4 = soup.find('div', id = 'column2') # یک ابجکت از نوع beautifulsoup
h3_soup = headline_bs4.findAll('h3')
news_urls =[]
for each_h3 in h3_soup:
    news_urls.append(each_h3.find('a').get('href'))
```

```python
with open ('iranjib.txt', 'w', encoding='utf-8') as file: #سیو کردن تو فایل 9
    file.write('id, title, summary,content, date, views\n')
    for id, page_url in enumerate(news_urls): #9
    #for page_url in news_urls:
        page_response = requests.get(page_url) #1
        page_soup = BeautifulSoup(page_response.text, 'html.parser')
        title = page_soup.find('h1').get_text()
        summary = page_soup.find('div',{'class' : 'newssummary'}).get_text() #2
        matn_soup = page_soup.find('div', {'class' : 'matn'}) #3
        paragraphs = matn_soup.find_all('p')
        news_text = ''  #4
        for p in paragraphs:
            news_text += p.get_text() + '\n'
        table_soup = page_soup.find_all('table')[1]   #5
        tr_soup = table_soup.find_all('tr')[1] #6
        date = tr_soup.find_all('td')[1].get_text() #7
        views = tr_soup.find_all('td')[2].get_text() #8
        file.write(f'{id}, {title}, {summary}, {news_text}, {date}, {views}\n')
```
